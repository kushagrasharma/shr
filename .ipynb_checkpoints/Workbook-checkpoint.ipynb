# Grouped analysis by ID

```python
by_id = df.groupby('id')
```

```python
revenue_by_id = by_id['revenue'].sum().sort_values(ascending=False)
revenue_by_id.describe()
```

```python
revenue_by_id.head(10)
```

```python
top_ids_revenue = revenue_by_id[top_ids.keys()]
sum(top_ids_revenue)/sum(revenue_by_id)
```

```python
profit_by_id = by_id['profit'].sum().sort_values(ascending=False)
profit_by_id.describe()
```

```python
profit_by_id.head(10)
```

```python
top_ids_profit = profit_by_id[top_ids.keys()]
sum(top_ids_profit)/sum(revenue_by_id)
```

```python
profit_deciles = pd.qcut(profit_by_id, 10)
```

```python
deciled_stats = defaultdict(lambda: [0, 0])
for index, value in profit_deciles.iteritems():
    deciled_stats[value][0] += revenue_by_id[index]
    deciled_stats[value][1] += profit_by_id[index]
df_deciles = pd.DataFrame(deciled_stats, index=['revenue', 'profit'])
df_deciles = df_deciles.transpose()
df_deciles
```

```python
ind = np.arange(10)
width = .25

fig, ax = plt.subplots()
rects1 = ax.bar(ind, df_deciles.revenue.values, width, color='r')

rects2 = ax.bar(ind+width, df_deciles.profit.values, width, color='b')

ax.set_xlabel("Profit decile")
ax.set_title('Revenue and profit by decile')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels([str(x) for x in range(1, 11)])
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

ax.legend((rects1[0], rects2[0]), ('Revenue', 'Profit'), loc=0)
```

```python
last_interval = df_deciles.revenue.keys()[-1]
top_decile = []
for index, value in profit_deciles.iteritems():
    if value == last_interval:
        top_decile.append(index)
```

```python
top_df = df.loc[df['id'].isin(top_decile)]
top_df.head(10)
```

```python
top_df.describe()
top_grouped = top_df.groupby('id')
```

```python
order_deciles = pd.cut(ids_count, [1, 2, 3, 4, 500])
order_deciles_count = order_deciles.value_counts()
order_deciles_count
```

```python
ind = np.arange(4)

#fig, ax = plt.subplots()
plt.bar(ind, order_deciles_count, color='r')

plt.xlabel("Number of orders")
plt.title('Order number frequency distribution')
plt.ylabel("Frequency")
plt.xticks(ind + .8 / 2, ['2', '3', '4', '5+'])
```

```python
o_deciled_stats = defaultdict(lambda: [0, 0])
for index, value in order_deciles.iteritems():
    o_deciled_stats[value][0] += revenue_by_id[index]
    o_deciled_stats[value][1] += profit_by_id[index]
o_df_deciles = pd.DataFrame(o_deciled_stats, index=['revenue', 'profit'])
o_df_deciles = o_df_deciles.transpose()
o_df_deciles
```

```python
ind = np.arange(4)
width = .25

fig, ax = plt.subplots()
rects1 = ax.bar(ind, o_df_deciles.revenue.values, width, color='r')

rects2 = ax.bar(ind+width, o_df_deciles.profit.values, width, color='b')

ax.set_xlabel("# of orders")
ax.set_title('Revenue and profit by # of orders')
ax.set_xticks(ind + width / 2)
ax.set_xticklabels(['2', '3', '4', '5+'])
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)

ax.legend((rects1[0], rects2[0]), ('Revenue', 'Profit'), loc=0)
```
