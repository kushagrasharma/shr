# Date-based Cohort Analysis on Entire Data Set

See http://www.gregreda.com/2015/08/23/cohort-analysis-with-python/ for details

```python
by_store = df.groupby('substore')
```

```python
top_ten_stores = repeats.substore.value_counts()[:10]
top_ten = df[df.substore.isin(top_ten_stores.index)]
top_ten.substore.cat.remove_unused_categories()
top_ten[top_ten.repeat_customer==True].drop_duplicates(subset='id').substore.value_counts()[:10]
```

```python
def cohortGroups(store):
    store['quarter'] = store.date.apply(lambda x: "{}-{}".format(x.year, ((x.month-1)//3)+1))
    store.set_index('id', inplace=True)
    store['CohortGroup'] = store.groupby(level=0)['date'].min().apply(lambda x: "{}-{}".format(x.year, 
                                                                                                ((x.month-1)//3)+1))
    store.reset_index(inplace=True)
    store = store[np.logical_not(store.CohortGroup.isin(['2007-4', '2009-1', '2008-1', '2008-3', '2009-4', '2009-2', '2009-3', '2008-4', '2016-3', '2016-4', '2017-1', '2017-2', '2017-3', '2010-1', '2010-2', '2010-3', '2010-4']))]
    return store
```

```python
def cohort_period(A):
    """
    Creates a `CohortPeriod` column, which is the Nth period based on the user's first purchase.
    
    Example
    -------
    Say you want to get the 3rd month for every user:
        df.sort(['UserId', 'OrderTime', inplace=True)
        df = df.groupby('UserId').apply(cohort_period)
        df[df.CohortPeriod == 3]
    """
    A['CohortPeriod'] = np.arange(len(A)) + 1
    return A
```

```python
def cohortAnalysis(df):
    cohort_df = cohortGroups(df).groupby(["CohortGroup", 'quarter']).agg({'id': pd.Series.nunique,
                       'date': pd.Series.nunique})
    cohort_df.rename(columns={'id': 'TotalUsers', 'date':'TotalOrders'}, inplace=True)
    cohort_df = cohort_df.groupby(level=0).apply(cohort_period)
    cohort_df = cohort_df[cohort_df.CohortPeriod <= 4]
    cohort_df.reset_index(inplace=True)
    cohort_df.set_index(['CohortGroup', 'CohortPeriod'], inplace=True)
    cohort_group_size = cohort_df['TotalUsers'].groupby(level=0).first()
    user_retention = cohort_df['TotalUsers'].unstack(0).divide(cohort_group_size, axis=1)
    return cohort_df, user_retention, cohort_group_size
```

```python
cohort_df, user_retention, cohort_group_size = cohortAnalysis(df)
```

```python
user_retention[['2011-1', '2012-1', '2013-1', '2014-1', '2015-1', '2016-1']].plot(figsize=(10,5))
plt.title('Cohort Analysis: User Retention')
plt.xticks(np.arange(1, 4.1, 1))
plt.xlim(1, 4)
plt.ylabel('% of Cohort Purchasing');
```

```python
sns.set(style='white')

plt.figure(figsize=(20, 14))
plt.title('Cohort Analysis (All Brands): User Retention')
sns.heatmap(user_retention.T, mask=user_retention.T.isnull(), annot=True, fmt='.3%');
```

# Cohort Analysis by Brand on Top 10 Brands

```python
def cohortSize(df):
    cohort_df = cohortGroups(df).groupby(["CohortGroup", 'quarter']).agg({'id': pd.Series.nunique,
                       'date': pd.Series.nunique})
    cohort_df.rename(columns={'id': 'TotalUsers', 'date':'TotalOrders'}, inplace=True)
    cohort_df = cohort_df.groupby(level=0).apply(cohort_period)
    cohort_df = cohort_df[cohort_df.CohortPeriod <= 4]
    cohort_df = cohort_df.groupby('CohortPeriod').agg({'TotalUsers':np.sum})
    return cohort_df
```

```python
def cohortByBrand(df):
    df = df.groupby('substore')
    t_df = pd.DataFrame(index=[1, 2, 3, 4])
    g = []
    for name, group in df:
        if group.empty:
            continue
        t_df[name] = cohortSize(group)
    user_retention = t_df.divide(t_df.iloc[0], axis=1)
    return t_df, user_retention
```

```python
brand_df, brand_retention = cohortByBrand(top_ten)
```

```python
brand_retention.plot(figsize=(10,5))
plt.title('User Retention By Brand - Cohort Analysis')
plt.xticks(np.arange(1, 4.1, 1))
plt.xlim(1, 4)
plt.ylabel('% of Cohort Purchasing');
```

```python
sns.set(style='white')

plt.figure(figsize=(15, 7))
plt.title('User Retention By Brand - Cohort Analysis')
sns.heatmap(brand_retention.T, mask=brand_retention.T.isnull(), annot=True, fmt='.3%');
```

# Retention By Purchase

```python
def customerRetention(store):
    if store.empty:
        return pd.Series(index=['1', '2', '3', '4', '5+'], data=[0,0,0,0,0])
    counts = store.id.value_counts().value_counts()
    data = [np.sum(counts)]
    for i in range(1, 5):
        data.append(data[i-1]-counts.iloc[i-1])
    return pd.Series(index=['1', '2', '3', '4', '5+'], data=data)
```

```python
purchase_count = customerRetention(df)
purchase_retention = purchase_count.divide(purchase_count.iloc[0])
```

```python
purchase_retention
```

```python
order_count = top_ten.groupby('substore').apply(customerRetention)
order_count = order_count[(order_count.T != 0).any()].T
c_retention = order_count.divide(order_count.iloc[0], axis=1)
c_retention['All Substores'] = purchase_retention
```

```python
c_retention.plot(figsize=(10,5))
plt.title('User Retention By Brands')
plt.xticks(np.arange(1, 4.1, 1))
plt.xlim(1, 4)
plt.ylabel('% of Cohort Purchasing');
```

```python
sns.set(style='white')

plt.figure(figsize=(15, 7))
plt.title('User Retention By Brand')
sns.heatmap(c_retention.T, mask=c_retention.T.isnull(), annot=True, fmt='.3%');
```

```python
def channelPath(r, len_path=3):
    r.sort_values('n_order')
    path = pd.Series(index=['initial_channel', 'c2', 'c3'], data=list(r.channel[:min(len_path,len(r))].values) + [np.NaN]*(len_path-len(r)))
    #path.cat.remove_unused_categories()
    return path
```

```python
paths = by_channel.apply(channelPath).dropna(thresh=1).reset_index(level=0)
paths.head(10)
```

```python
init_counts = paths.initial_channel.value_counts()
i_total = np.sum(paths.initial_channel.value_counts())
initial_dist = pd.DataFrame(columns=['initial_channel', 'distribution'],
                            data=[[name, value/i_total] for name, value in init_counts.iteritems()]).set_index('initial_channel')
initial_dist.head()
```

```python
def secondChannelDistribution(channel):
    total = float(np.sum(channel.values))
    dist = [[name[1], value/total] for name, value in channel.c2.iteritems()]
    return pd.DataFrame(columns=['c2', 'distribution'], data=dist).set_index('c2')
```

```python
top_by_channel = paths.groupby('initial_channel')
second_channel = pd.DataFrame(top_by_channel.apply(lambda x: x.c2.value_counts()))
second_dist = second_channel.groupby('initial_channel').apply(secondChannelDistribution)
second_dist.head()
```

```python
def thirdChannelDistribution(channel):
    total = float(np.sum(channel.values))
    dist = [[name[2], value/total] for name, value in channel.c3.iteritems()]
    return pd.DataFrame(columns=['c3', 'distribution'], data=dist).set_index('c3')
```

```python
third_channel = pd.DataFrame(paths.groupby(['initial_channel', 'c2']).apply(lambda x: x.c3.value_counts()))
third_dist = third_channel.groupby(['initial_channel', 'c2']).apply(thirdChannelDistribution)
third_dist.head(5)
```

```python
dist = third_dist.rename(columns = {'distribution':'third_channel_dist'}).reset_index()
dist['first_channel_dist'] = [float(initial_dist.loc[chan]) for chan in dist.initial_channel]
dist['second_channel_dist'] = [float(second_dist.loc[row.initial_channel, row.c2]) for _, row in dist.iterrows()]
dist.head()
```

```python
cumulative_dist = dist.copy()
cumulative_dist.second_channel_dist *= cumulative_dist.first_channel_dist
cumulative_dist.third_channel_dist *= cumulative_dist.second_channel_dist
cumulative_dist.head()
```

```python
x = second_dist.loc[[x for x, y in channel_type.items() if y in ['PPC (Main)', 'PPC (Other)']]]#.sort_values('distribution', ascending=False)
x.distribution = [row.distribution*initial_dist.loc[index[0]].distribution for index, row in x.iterrows()]
x=x.reset_index(level=0, drop=True).reset_index().groupby('c2').agg(np.sum).sort_values('distribution', ascending=False)
```
